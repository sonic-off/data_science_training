```python
import numpy as np
import pandas as pd
```

```python
metal_df = pd.read_csv("2020-06-18_metal_temp.csv")
```

```python
metal_df
```

```python
metal_df.plot.scatter(x='metal', y='gdp')
```

```python
# sklearn.linear_model.LinearRegression クラスを読み込み
from sklearn import linear_model
metal = linear_model.LinearRegression()
```

```python
# データの対数変換と2Darrayへの加工
x_ = np.log(metal_df.metal)
y_ = np.log(metal_df.gdp)
x = x_.values
y = y_.values
X = x.reshape(1, -1)
Y = y.reshape(1,-1)
```

```python
from sklearn.linear_model import LinearRegression
lr = linear_model.LinearRegression(fit_intercept=True, normalize=False, copy_X=True, n_jobs=1)
lr.fit(X, Y)
```

```python
# 回帰係数。本来は1個で出るはずだがなぜか0.0.0…
lr.coef_
```
